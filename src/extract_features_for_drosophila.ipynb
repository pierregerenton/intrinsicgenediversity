```python
import pandas as pd
```

```python
complete_fly_annotation = pd.read_csv('../data/Drosophila_melanogaster.BDGP6.46.111.gff3', sep='\t', header=None, comment='#')
```

```python
complete_fly_annotation
```

```python
# keeping usefull line
feature_fly_annotation = complete_fly_annotation[complete_fly_annotation[2].isin(['gene', 'mRNA', 'exon', 'CDS'])]
```

```python
# keeping usefull column
fly_annotation = feature_fly_annotation.loc[:,[0,2,3,4,8]]
```

```python
fly_annotation
```

```python
gene = fly_annotation[fly_annotation[2].isin(['gene'])]
gene
```

```python
only_protein_coding = gene[8].apply(lambda x : "biotype=protein_coding" in x)
gene = gene[only_protein_coding]
with_readthrough = gene[8].apply(lambda x : "readthrough" in x)
gene = gene[~with_readthrough]
gene['ID'] = gene[8].apply(lambda x : x.split(';')[0].lstrip('ID='))
coding_gene = set(gene['ID'].tolist())
gene
```

```python
def infogff2dict(x):
    return dict( (k,v) for k,v in [element.split('=') for element in x.split(';')])
```

```python
mRNA = fly_annotation[fly_annotation[2].isin(['mRNA'])]
mRNA
```

```python
mRNA_from_coding_gene = mRNA[8].apply(lambda x : infogff2dict(x)['Parent'] in coding_gene)
mRNA[mRNA_from_coding_gene]
```

```python
only_protein_coding = mRNA[8].apply(lambda x : infogff2dict(x)['biotype'] == 'protein_coding')
mRNA = mRNA[only_protein_coding]
mRNA['ID'] = mRNA[8].apply(lambda x : x.split(';')[0].lstrip('ID='))
coding_mRNA = set(mRNA['ID'].tolist())
mRNA
```

```python
cds = fly_annotation[fly_annotation[2].isin(['CDS'])]
print(cds[8][505])
```

```python
# Making transcript -> gene dictionnary
tr2gene = dict()
for index, row in mRNA.iterrows():
    row = list(row)
    info = infogff2dict(row[4])
    gene_id : str = info['Parent']    
    tr_id : str = info['ID']
    tr2gene[tr_id] = gene_id
```

```python
# Making a dictionnary with gene -> transcript -> cds structure
repartition = dict() 
for index, row in cds.iterrows():
    row : list = list(row)
    chr : str = row[0]
    start : int = int(row[2])
    end : int = int(row[3])
    info = infogff2dict(row[4])
    transcript_id : str = info['Parent']
    gene_id = tr2gene[transcript_id]
    if gene_id in repartition:
        if transcript_id in repartition[gene_id]:
            repartition[gene_id][transcript_id].add((chr, start, end))
        else:
            repartition[gene_id][transcript_id] = {(chr, start, end)}
    else:
        repartition[gene_id] = {transcript_id : {(chr, start, end)}}
```

```python
repartition
```

```python
print(len(repartition))
n_isoform_no_filter = [len(repartition[k]) for k in repartition]
sum(n_isoform_no_filter)
```

```python
repartition_no_dup : dict[str, dict[str, set[tuple[str, int, int]]]] = dict() 
for gene in repartition:
    repartition_no_dup[gene] = dict()
    for transcript in repartition[gene]:
        if repartition[gene][transcript] not in repartition_no_dup[gene].values():
            repartition_no_dup[gene][transcript] = repartition[gene][transcript]
```

```python
print(len(repartition_no_dup))
n_isoform_no_filter = [len(repartition_no_dup[k]) for k in repartition_no_dup]
sum(n_isoform_no_filter)
```

```python
coding_gene = coding_gene
coding_mRNA = []
for ge in repartition_no_dup.values():
    coding_mRNA.extend(list(ge.keys()))
print(len(coding_mRNA))
coding_mRNA = set(coding_mRNA)
```

```python
def keeprow(x):
    if x[2] == 'gene':
        return infogff2dict(x[8])['ID'] in coding_gene
    elif x[2] == 'mRNA':
        return infogff2dict(x[8])['ID'] in coding_mRNA
    elif x[2] == 'exon':
        return infogff2dict(x[8])['Parent'] in coding_mRNA
    elif x[2] == 'CDS':
        return infogff2dict(x[8])['Parent'] in coding_mRNA
    else:
        False
```

```python
row_kept = feature_fly_annotation.apply(keeprow, axis=1)
```

```python
feature_fly_annotation = feature_fly_annotation[row_kept]
```

```python
feature_fly_annotation.to_csv('../data/flybase.dm.coding_gene.coding_transcript.gff', header=None, sep='\t', index=False)
```
